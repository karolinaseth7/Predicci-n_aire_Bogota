```python
# # Rutas de tus archivos
# ruta_ubicaciones = r"D:\Maestria\Series de tiempo\Data\ubicaci√≥n.xlsx"
# df_ubicacion = pd.read_excel(ruta_ubicaciones)

# Mostrar nombres exactos
print("Columnas en df_ubicacion antes de renombrar:")
print(df_ubicacion.columns.tolist())

# --- Normalizar nombres ---
df.columns = df.columns.str.strip().str.lower()
df_ubicacion.columns = df_ubicacion.columns.str.strip().str.lower()

# Si tus columnas se llaman distinto, ajusta aqu√≠
df_ubicacion = df_ubicacion.rename(columns={
    'latitud': 'latitud_decimal',
    'longitud': 'longitud_decimal'
})

# Buscar columnas de estaci√≥n
col_estacion_df = [c for c in df.columns if 'estacion' in c][0]
col_estacion_ubi = [c for c in df_ubicacion.columns if 'estacion' in c][0]

# Unir DataFrames
df = df.merge(
    df_ubicacion[[col_estacion_ubi, 'latitud_decimal', 'longitud_decimal']],
    left_on=col_estacion_df,
    right_on=col_estacion_ubi,
    how='left'
)

# Eliminar columna duplicada si existe
if col_estacion_ubi != col_estacion_df and col_estacion_ubi in df.columns:
    df.drop(columns=[col_estacion_ubi], inplace=True)

# Guardar resultado
ruta_salida = r"D:\Maestria\Series de tiempo\Data\serie_tiempo_estaciones_con_coordenadas.csv"
df.to_csv(ruta_salida, index=False)

print("\n‚úÖ Uni√≥n completada correctamente.")
print(f"üìÅ Archivo guardado en: {ruta_salida}")
print("\nEjemplo de datos con coordenadas:")
print(df[[col_estacion_df, 'latitud_decimal', 'longitud_decimal']].head())
```
